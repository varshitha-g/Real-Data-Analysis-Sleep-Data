```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import ttest_rel, wilcoxon, shapiro, norm
```

```python
# Simulated Sleep Data
np.random.seed(42)
sleep_data = pd.DataFrame({
    "Group1": np.random.normal(loc=7, scale=1.5, size=10),  # Sleep hours for Group 1
    "Group2": np.random.normal(loc=8, scale=1.5, size=10)   # Sleep hours for Group 2
})
sleep_data["Difference"] = sleep_data["Group2"] - sleep_data["Group1"]
```

```python
# Descriptive statistics
print("Descriptive Statistics:")
print(sleep_data.describe())
```

```python
# Parametric Test: Paired t-test
t_stat, t_p_value = ttest_rel(sleep_data["Group1"], sleep_data["Group2"])
print(f"Paired t-test: t-statistic = {t_stat}, p-value = {t_p_value}")
```

```python
# Non-parametric Test: Wilcoxon signed-rank test
w_stat, w_p_value = wilcoxon(sleep_data["Group1"], sleep_data["Group2"])
print(f"Wilcoxon signed-rank test: w-statistic = {w_stat}, p-value = {w_p_value}")
```

```python
# Normality Test for Differences
shapiro_stat, shapiro_p = shapiro(sleep_data["Difference"])
print(f"Shapiro-Wilk Test for Normality: stat = {shapiro_stat}, p-value = {shapiro_p}")
```

```python
# Box Plot for Group Comparison
plt.figure(figsize=(10, 6))
plt.boxplot([sleep_data["Group1"], sleep_data["Group2"]], labels=["Group 1", "Group 2"])
plt.title("Box Plot of Sleep Hours by Group")
plt.ylabel("Sleep Hours")
plt.grid()
plt.savefig("box_plot_groups.png")  # Save plot
plt.show()
```

```python
# Normal Plot of Differences
plt.figure(figsize=(10, 6))
plt.hist(sleep_data["Difference"], bins=10, alpha=0.7, label="Differences", color='blue')
plt.title("Distribution of Differences in Sleep Hours")
plt.xlabel("Difference in Sleep Hours")
plt.ylabel("Frequency")
plt.grid()
plt.legend()
plt.savefig("histogram_differences.png")  # Save plot
plt.show()
```

```python
# Simulating Power Function
effect_sizes = np.linspace(0.1, 1.0, 10)
sample_size = 10
simulations = 1000
parametric_power = []
nonparametric_power = []

for effect in effect_sizes:
    parametric_rejections = 0
    nonparametric_rejections = 0
    
    for _ in range(simulations):
        group1 = np.random.normal(7, 1.5, sample_size)
        group2 = np.random.normal(7 + effect, 1.5, sample_size)
        
        # Paired t-test
        _, p_param = ttest_rel(group1, group2)
        if p_param < 0.05:
            parametric_rejections += 1
        
        # Wilcoxon signed-rank test
        _, p_nonparam = wilcoxon(group1, group2)
        if p_nonparam < 0.05:
            nonparametric_rejections += 1
    
    parametric_power.append(parametric_rejections / simulations)
    nonparametric_power.append(nonparametric_rejections / simulations)
```

```python
# Plot Power Functions
plt.figure(figsize=(10, 6))
plt.plot(effect_sizes, parametric_power, label='Parametric (t-test)', marker='o')
plt.plot(effect_sizes, nonparametric_power, label='Non-Parametric (Wilcoxon)', marker='x')
plt.title('Power Function: Parametric vs Non-Parametric Tests')
plt.xlabel('Effect Size')
plt.ylabel('Power')
plt.legend()
plt.grid()
plt.savefig("power_function_plot.png")  # Save plot
plt.show()
```
